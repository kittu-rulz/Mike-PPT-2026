# Extract Slide 5 Logos from PPTX

This notebook inspects slide 5 of the attached PowerPoint file and extracts picture shapes that appear to be logos. It saves each found image locally for verification.


```python
from pathlib import Path
from PIL import Image
from pptx import Presentation

pptx_path = Path('ENT - FY’26 Performace & FY’27 Growth Plan-02-Apr.pptx')
output_dir = Path('extracted-slide5-logos')
output_dir.mkdir(exist_ok=True)

print('PPTX exists:', pptx_path.exists())
print('Output dir:', output_dir.resolve())
```

## Load the PowerPoint File

Open the presentation object and inspect the total number of slides.


```python
presentation = Presentation(str(pptx_path))
print('Total slides:', len(presentation.slides))
slide_index = 4  # slide 5 is index 4
slide = presentation.slides[slide_index]
print('Loaded slide index:', slide_index, 'title:', getattr(slide.shapes.title, 'text', 'No title'))
```

## Inspect Slide 5 Shapes

List all shapes on slide 5, with type and image metadata when available.


```python
from pptx.enum.shapes import MSO_SHAPE_TYPE

for i, shape in enumerate(slide.shapes, start=1):
    shape_type = shape.shape_type
    shape_name = shape.name
    print(f'Shape {i}:', shape_name, 'type=', shape_type)
    if shape_type == MSO_SHAPE_TYPE.PICTURE:
        print('  Picture: width=', shape.width, 'height=', shape.height)
        print('  image content type=', shape.image.content_type)
        print('  filename=', shape.image.filename)
    elif shape.has_text_frame:
        print('  text=', shape.text)
    print('---')
```

## Extract and Save Slide 5 Logo Images

Save any picture shapes found on slide 5 to the output directory.


```python
saved_files = []
for i, shape in enumerate(slide.shapes, start=1):
    if shape.shape_type == MSO_SHAPE_TYPE.PICTURE:
        image = shape.image
        ext = Path(image.filename).suffix or '.png'
        out_path = output_dir / f'slide5-logo-{i}{ext}'
        with open(out_path, 'wb') as f:
            f.write(image.blob)
        saved_files.append(out_path)
        print('Saved:', out_path)

print('Total images saved:', len(saved_files))
```

## Display Extracted Logo Images

Show the saved images inline for review.


```python
from IPython.display import display
for image_path in saved_files:
    img = Image.open(image_path)
    print(image_path.name)
    display(img)
```
